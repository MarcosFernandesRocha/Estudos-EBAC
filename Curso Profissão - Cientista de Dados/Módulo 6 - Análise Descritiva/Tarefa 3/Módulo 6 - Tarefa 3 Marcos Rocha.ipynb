```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df_rn_ro = pd.read_csv('SINASC_RO_2019_versao2.csv')

df_rn_ro.info()

df_rn_ro
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df_rn_ro.loc[(df_rn_ro['apgar5'] >= 0) & 
              (df_rn_ro['apgar5'] <= 3), 
              'class_apgar5'] = 'asfixia grave'

df_rn_ro.loc[(df_rn_ro['apgar5'] >= 4) & 
              (df_rn_ro['apgar5'] <= 5), 
              'class_apgar5'] = 'asfixia moderada'

df_rn_ro.loc[(df_rn_ro['apgar5'] >= 6) & 
              (df_rn_ro['apgar5'] <= 7), 
              'class_apgar5'] = 'asfixia leve'

df_rn_ro.loc[(df_rn_ro['apgar5'] >= 8) & 
              (df_rn_ro['apgar5'] <= 10), 
              'class_apgar5'] = 'normal'

df_rn_ro['class_apgar5'].value_counts()
```

```python
#Checando a frequência da cluna de gestação
df_rn_ro['gestacao'].value_counts()
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
#Gráfico com a média de APGAR5 por semanas de gestação
sns.catplot(x='gestacao', 
            y='apgar5',
            hue='sexo', 
            data=df_rn_ro, 
            order=['Menos de 22 semanas', 
                   '22 a 27 semanas', 
                   '28 a 31 semanas', 
                   '32 a 36 semanas', 
                   '37 a 41 semanas', 
                   '42 semanas e mais'], 
            kind='bar')
plt.xticks(rotation=90)
plt.xlabel(xlabel='Tempo de gestação')
plt.ylabel(ylabel='Média de "APGAR5"')

plt.show()
```

```python
#Outra forma de visualizar com histogramas separando por diferentes semanas de gestação:
tempo_gestacao = df_rn_ro['gestacao'].value_counts().reindex(
    ['Menos de 22 semanas', 
     '22 a 27 semanas', 
     '28 a 31 semanas', 
     '32 a 36 semanas', 
     '37 a 41 semanas', 
     '42 semanas e mais']).index

for i in tempo_gestacao:
    sns.displot(data = df_rn_ro[df_rn_ro.gestacao == i ]
            ,x = 'apgar5'
            ,hue = 'sexo'
            ,bins = 20
            ,element = 'step'
           )
    plt.legend(loc='upper left')
    plt.title(label='Tempo de gestação: {}'.format(i))
    plt.show()
```

```python
#Outra forma de visualizar os dados relacionando o APGAR5 às semanas de gestação
sns.catplot(x='sexo', y='apgar5',
            hue='gestacao',
            kind='bar',
            data=df_rn_ro,
            errorbar=None)
            
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python

tabela_apgar5 = pd.crosstab(index=df_rn_ro['gestacao'], columns=df_rn_ro['class_apgar5'])

tabela_apgar5 = tabela_apgar5.drop('9')

tabela_apgar5


```

```python
#Somando os dados obtidos por variável de semanas de gestação
print(tabela_apgar5.sum(1))
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
#Percentuais de cada classe de gestação:
freq_tabela_apgar5  = tabela_apgar5.div(tabela_apgar5.sum(axis = 1), axis = 0)
print(freq_tabela_apgar5)
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
#Gráfico editado a partir dos dados percentuais acima
plt.rc('figure', figsize=(8,4))
freq_tabela_apgar5.plot.bar()
plt.xticks(rotation=45)
plt.xlabel(xlabel='Tempo de gestação')
plt.ylabel(ylabel='Percentual')
plt.legend(title='Apgar no quinto minuto',fontsize=5)
plt.legend(loc='upper left')
```

##### Nota: De acordo com os dados do gráfico as complicações respiratórias começam a ser mais frequêntes quando temos partos com menos de 37 semanas.

#### 7) Veja se há relação entre APGAR5 e o peso do bebê. Isso pode ser feito de diversas formas.
```python
#Histograma de relação entre APGAR5 e peso do bebê
plt.rc('figure', figsize=(8,4))


ax1 = sns.barplot(x='apgar5', y='peso', data=df_rn_ro)
ax1.set_title(label='Relação APGAR5 X Peso do bebê (barras)')


```

```python
#Gráfico boxplot da relação entre APGAR5 e peso do bebê
ax2 = sns.boxplot(x='apgar5', y='peso', data=df_rn_ro)
ax2.set_title(label='Boxplot APGAR5 X Peso do bebê')
```

```python
#Relação de APGAR5 , peso do bebê e semanas de gestação 
sns.jointplot(data=df_rn_ro, x='apgar5', y='peso',hue = 'gestacao',alpha = 0.5)
plt.legend(loc='best',fontsize=7)

```

Ao observar as variáveis é possível perceber que existe relação entre o peso do bebê, APGAR5 e as semanas de gestação. Bebês com semanas de gestação 
menores apresentam pesos menores e valores de APGAR5 quase sempre inferiores a 10. Há também uma relação entre baixos pesos e semanas de gestação menores. Porém se o APGAR5 apresenta valores próximos a 10 também é possivel notar que o peso do bebê não tem tanta variação em relação aos valores mais baixos de APGAR5.
