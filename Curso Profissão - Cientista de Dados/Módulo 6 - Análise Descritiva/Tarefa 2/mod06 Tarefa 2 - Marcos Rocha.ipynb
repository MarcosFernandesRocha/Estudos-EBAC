```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
#importando a base de dados do arquivo CSV
df_rn_boxplot = pd.read_csv('SINASC_RO_2019_versao2.csv')
```

```python
print(df_rn_boxplot.columns)
```

```python
plt.rc('figure', figsize=(10,8))
fig, axes = plt.subplots(1,1)
sns.boxplot(y='idade_mae', data=df_rn_boxplot, ax=axes)
axes.set_title(label='Idade da Mãe')
axes.set_ylabel(ylabel='Idade da mãe (em anos)')

plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.rc('figure', figsize=(10,8))
fig, axes = plt.subplots(1,1)

sns.boxplot(y='peso', data=df_rn_boxplot, ax=axes)
axes.set_title(label='Peso do bebê')
axes.set_ylabel(ylabel='Peso ao nascer (em gramas)')

plt.show()
```

####  3) Faça um histograma da idade da mãe

```python

sns.displot(data = df_rn_boxplot['idade_mae'],alpha = .25,element = 'step',  kde = True, bins=10, ax=axes)

plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
sns.displot(data = df_rn_boxplot['peso'],alpha = .25,element = 'step',  kde = True, bins=10, ax=axes)

plt.show()
```
