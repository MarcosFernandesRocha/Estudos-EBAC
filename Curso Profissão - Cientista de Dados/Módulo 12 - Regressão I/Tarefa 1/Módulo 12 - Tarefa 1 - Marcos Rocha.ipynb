### 2. Mudança de perspectiva

Agora ajuste um modelo de regressão de ```tip_pct``` por ```net_bill```. Construa o gráfico de dispersão correspondente.

````python
reg_tip_pct = smf.ols(formula='tip_pct ~ net_bill', 
                      data=tips
                     ).fit()

reg_tip_pct.summary()
````

````python
# Gráfico de dispersão
sns.regplot(y = 'tip', x = 'tip_pct', data = tips)

# Gerando dados para a linha dos 10%
x = np.linspace(0,tips['tip_pct'].max(),100)
y = .1*x

# Plotando a linha dos 10%
plt.plot(x, y, '-r', label='linha dos 10%')

# Opções estéticas para explicitar o (0,0) no gráfico
plt.xlim(left = 0)
plt.ylim(bottom = 0)
````

````python
tips['tip_pct_predict'] = reg_tip_pct.predict(tips)
````

````python
tips[['net_bill','tip','tip_predict','tip_pct','tip_pct_predict']]
````

````python
# Gráfico de dispersão
sns.regplot(y = 'tip', x = 'tip_pct_predict', data = tips)

# Gerando dados para a linha dos 10%
x = np.linspace(0,tips['tip_pct_predict'].max(),100)
y = .1*x

# Plotando a linha dos 10%
plt.plot(x, y, '-r', label='linha dos 10%')

# Opções estéticas para explicitar o (0,0) no gráfico
plt.xlim(left = 0)
plt.ylim(bottom = 0)
````
