#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
df_last60 = df.xs(key='AMER3.SA', 
                  level=1, 
                  axis=1).tail(n=60)

print(f'Quantidade de linhas do período selecionado: {df_last60.shape[0]}.\n')
df_last60
````

````python
layout = Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)

graph = {
    'x': df_last60.index, 
    'open': df_last60.Open,
    'close': df_last60.Close,
    'high': df_last60.High,
    'low': df_last60.Low,
    'type': 'candlestick',
    'name': 'AMER3.SA',
    'showlegend': True
}

grafico = graph_objects.Figure(data=[graph], 
                               layout_title='Valor de Mercado Americanas SA.', 
                               layout=layout)

grafico.update_xaxes(showgrid=True, 
                     gridwidth=1, 
                     gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True, 
                     gridwidth=1, 
                     gridcolor='LightGrey')

grafico
````
